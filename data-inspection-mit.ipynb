```python
import wfdb
import pandas as pd
import numpy as np

import os
import matplotlib.pyplot as plt
```

```python
DATA_FOLDER = 'data/mit-bih-arrhythmia-database-1.0.0/'
```

# MIT ECG

```python
from typing import List
def show_ecg_signals(record_name: str, channels: List[int] = None, sampto: int = None):
    record = wfdb.rdrecord(DATA_FOLDER + record_name, channels=channels, sampto=sampto)
    annotation = wfdb.rdann(DATA_FOLDER + record_name, 'atr')

    wfdb.plot_wfdb(record=record, title=f'Record of {record_name}')
    display(record.__dict__)
    display(annotation.__dict__)
    return record, annotation

record_name = '100'
record, annotation = show_ecg_signals(record_name)
```

## Get segmented results in 10 seconds

```python
record_name = '100'
duration_seconds = 10

# Calculate the corresponding sample numbers for the desired duration
sampling_frequency = record.fs  # Sampling frequency in Hz
duration_samples = int(duration_seconds * sampling_frequency)

seg_record, seg_annot = show_ecg_signals(record_name, channels=[0], sampto=duration_samples)
```

# Dividing the ECG

## Get 46 signals ML II

```python
# Tentukan folder path tempat Anda memiliki file .dat
folder_path = DATA_FOLDER

# Dapatkan semua nama file .dat dari folder
dat_files = [f for f in os.listdir(folder_path) if f.endswith('.dat')]

# Ambil nama file tanpa ekstensi
file_names_without_extension = [os.path.splitext(f)[0] for f in dat_files]

print(file_names_without_extension)
print(len(file_names_without_extension))
```

```python
records = []
informations = []

# Membaca sinyal MLII dari semua rekaman
for file_name in file_names_without_extension:
    record, information = wfdb.rdsamp(folder_path + file_name, channel_names=['MLII'])
    if record is None:
        continue

    records.append(record)
    informations.append(information)

    print(f'Record {file_name} contains {len(record)}')

print(informations[0])
len(records)
```

```python
print(records[0])
```

## Divide to two pieces of 10s

```python
def divide_to_two_10s(record):
    return np.array(record[:duration_samples]).flatten(), np.array(record[duration_samples:2*duration_samples]).flatten()

first_record_divided = divide_to_two_10s(records[0])
print(len(first_record_divided[0]))
print(first_record_divided[0])
print(len(first_record_divided[1]))
```

## Create a table training

```python
# examples of algorithm for two consecutives left and right
data = list(range(1, 3601))
print(data)

# algoritma untuk mengambil susunan i, i+1, i+3, i+4, i+2 dari data
def create_consecutives_data(used_data, phase: int = 1):
    selected_data = []
    starting_point = phase - 1

    while starting_point + 5 < len(used_data):
        consecutive_group_indexes = [0, 1, 3, 4, 2]
        selected_datum = [used_data[starting_point + j] for j in consecutive_group_indexes]
        selected_data.append(selected_datum)
        starting_point += 3
    
    return selected_data

selected_data = create_consecutives_data(data, 1)

print(selected_data[0])
print(selected_data[-1])
print(len(selected_data))
```

```python
training_records = []
for record in records:
    # Mendapatkan value 10s untuk training dan testing
    divided_record = divide_to_two_10s(record)
    training_record = divided_record[0]

    # Ambil hasil pengolahan consecutive masukkan ke list training
    selected_training_data = create_consecutives_data(training_record)
    training_records += selected_training_data

# Mengubah array 2D menjadi DataFrame
df = pd.DataFrame(training_records)
print(df.shape)

# Menyimpan DataFrame ke dalam file CSV
df.to_csv('training.csv', index=False, header=False)
```
